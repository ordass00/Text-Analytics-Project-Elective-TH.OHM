Erklärung zum Spiel:

Cyberpunk 2077 ist ein Open-World-Action-Adventure, das in der virtuellen Stadt Night City spielt. Es war vor seinem Release, laut einer Umfrage, das meist erwartete Spiel ([Quelle](https://www.cdprojekt.com/en/media/news/cyberpunk-2077-claims-this-years-golden-joystick-for-most-wanted-game/)). Das Spiel bekam zum Release sehr viele gute Wertungen von verschiedenen Spielemagazinen ([Quelle](https://www.metacritic.com/game/pc/cyberpunk-2077)). Drauffolgend gab es aber sehr viele negative Schlagzeilen und ein Spielemagazin gab zu, einen Fehler bei der Bewertung des Spiels gemacht zu haben
([Quelle_1](https://www.gamestar.de/artikel/cyberpunk-2077-savegame-kaputt,3365344.html), [Quelle_2](https://www.gamestar.de/artikel/cyberpunk-2077-ps4-xbox-technik-meinung,3365093.html), [Quelle_3](https://www.gamestar.de/artikel/cyberpunk-2077-sammelklage-cd-projekt,3365389.html), [Quelle_4](https://www.gamestar.de/artikel/cyberpunk-2077-report-kritik-reaktion,3366043.html), [Quelle_5](https://www.gamestar.de/artikel/cyberpunk-2077-entschuldigung-patch-105,3365123.html), [Quelle_6](https://www.gamestar.de/artikel/cyberpunk-2077-abwertung,3365051.html)).

 * Falls dies der Fall ist, möchte die Firma erfahren, wo die Probleme liegen, damit Sie passende Lösungen finden kann.

# Zusammenfassung: #
Zusammenfassend kann man sagen, dass sich die anfangs befürchtete negative Einstellung der Spieler gegenüber dem Spiel im Lauf der Analyse immer weiter verstärkt hat. So wurden besonders häufig Wörter wie `bug`, `problem`, `issue` und `refund` allgegenwertig in dem Cyperpunk Subreddit gefunden und nur selten über positive Aspekte wie die `story` und die `character` gesprochen. Die negativen Aspekte haben bei der [Analyse der Worthäufigkeiten](./02_Explorative_Datenanalyse.ipynb#WordCloud) deutlich überwogen.


Trotz alldem wurde deutlich, dass Cyperpunk 2077 einen enormen Aufwind erleben hat dürfen. Dies zeigte sich nicht nur an positiven Beispielen wie der [steigenden Anzahl an Kommentaren zu Release](./02_Explorative_Datenanalyse.ipynb#Analyse-der-zeitlichen-Entwicklung-der-Kommentare), sondern eben auch zu negativen Ereignissen wie die Verschiebung des Releases vom 27.10.2020 zu 10.12.2020.


Das sich die negative Grundeinstellung bewahrheitet hat, wurde letztlich nochmal in der [Sentiment Analyse](./03_Sentiment_und_Topic_Analyse.ipynb#Sentiment-Analyse-Plotten) bestätigt. So wurden von knapp 442.000 Kommentaren gut 1/3 (~166.000) als negativ eingestuft und nur ca. 80.000 als positiv, wobei sich die negativen Kommentare in der [Topic Analyse](./03_Sentiment_und_Topic_Analyse.ipynb#Roberta-Topic-Analyse) in Themen wie `game breaking bug` und `run like shit` widergespiegelt haben.

Mithilfe der Topic Analyse konnten wir die Themen der negativen Kommentaren herausfinden. Neben den bereits vorher analysierten Themen gab es zwei neue Themen. Das eine Thema betraf die verschiedenen Versionen des Spiels. Das andere Thema, welches wir näher untersucht haben, bezieht sich auf das Gameplay des Spiels.










    
    

# Handlungsempfehlung: #













Man hat anhand der Anzahl der Kommentare gesehen, dass das Spiel ein großes Potential hatte. Wir empfehlen deswegen der Firma, das Spiel nicht aufzugeben und an den Problemen zu arbeiten. 

Im Detail heißt das, dass die verschiedenen Versionen des Spiels verbessert werden sollten. Die vielen Bugs, wie z.B. dass beim „character creation screen“ das Spiel abstürzt, sollten unbedingt behoben werden. Die Probleme hatten wir, wie bereits erwähnt, näher untersucht und festgestellt, dass es der „open world“ an Interaktionen mangelt. Außerdem muss die AI der NPC, Polizei und des Straßenverkehrs, um einiges verbessert werden. 

Bei all diesen Maßnahmen muss die Firma bewerten, ob diese im Bereich des Möglichen sind und die Spieler am besten informieren, was man als Nächstes plant.
